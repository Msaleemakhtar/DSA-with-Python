## Linked List -Practice

```python
# created single node 

class Node:
    def __init__(self, value):
        self.data = value
        self.next = None

class LinkedList:
    def __init__(self) -> None:
        # initialized empty linkedlist
        self.head = None
        # no of nodes
        self.n = 0

    def __len__(self):
        return self.n
    
    # print the ll 
    def __str__(self):
        current = self.head
        result = ""
        while current != None:
            result = result + str(current.data) + "-->"
            current = current.next
            

        return result[:-3]
        
    # insert from head 
    def insert_head(self, value):
        # created new node
        new_node = Node(value)
        #create connection
        new_node.next = self.head
        # reassign
        self.head = new_node
        #increament n
        self.n = self.n + 1

    # inseart from tail
    def append(self, value):
        new_node = Node(value)
        if self.head == None:
            self.head = new_node
            self.n = self.n + 1
            return

        current = self.head
        while current.next != None:
            current = current.next

        current.next = new_node
        self.n = self.n + 1

    # insert in between the linked list   
    def insert_after(self, after, value):
        new_node = Node(value)

        current = self.head
        while current != None:
            if current.data == after:
                break
            current =current.next
        
        if current != None:
            new_node.next = current.next
            current.next = new_node
            self.n = self.n + 1
        else: 
            return "Node not found"
        
    # clear the link list
    def clear(self):
        self.head = None
        self.n = 0


    # delete from head  
    def delete_head(self):
        if self.head == None:
            return "Linked List is Empty"
        self.head = self.head.next
        self.n = self.n - 1

    # delete from tail
    def pop(self):
        # if empty ll
        if self.head == None:
            return "Empty LL"
        
        current = self.head
        # if only one item in ll and that is head
        if current.next == None:
            return self.delete_head()
        # otherwise
        while current.next.next != None:
            current = current.next
        # cuurrnt --> 2nd last node
        current.next = None
        self.n = self.n - 1

    #delete from value
    def remove(self, value):
        
        if self.head == None:
            return "LL Empty"

        if self.head.data == value:
            return self.delete_head()

        current = self.head
        
        while current.next != None:
            if current.next.data == value:
                break
            current= current.next
        
        if current.next == None:
            return "Value not Found"
        else:
            current.next = current.next.next
            
    # search the index/position of elements in Linked List
    def search(self, value):
        current = self.head
        pos = 0

        while current != None:
            if current.data == value:
                return pos
            current = current.next
            pos = pos + 1

        return "Out of range"
    
    # search the value / item by index
    def __getitem__(self, index):

        current = self.head
        pos = 0

        while current != None:
            if pos == index:
                return current.data
            current = current.next
            pos += 1

        return "IndexError"
    
    
    # replace max value 
    def replace_max(self, value):
        current = self.head
        print("ist current : ", current)

        max = current
        print("max : ", max)

        while current != None:
            if current.data > max.data:
                print("if current.data : ", current.data)
                print("if max.data : ", max.data)
                max = current
                print("2nd current : ", current)  
                print("2max : ", max)  
            current = current.next
        print("last max.data : ", max.data)    
        max.data = value




        # sum the value of nodes at odd position
    def sum_odd_node(self):
        current = self.head
        counter = 0
        result = 0

        while current != None:
            if counter % 2 != 0:
                result = result + current.data
            counter +=1
            current = current.next
        print(result)

```
